# Time Series Analysis of Ford(F)

```python
## Computing Daily Returns

f_close_df["Daily Return"] = f_close_df["F_Close"].pct_change()

#Display the first few rows
f_close_df[['F_Close', 'Daily Return']].head()
```

```python
# Summary statistics
summary_stats = f_close_df['F_Close'].describe()
summary_stats
```

```python
f_close_df
```

```python
# Fill missing values using backward fill
f_close_df = f_close_df.fillna(method='bfill')

# Check if missing values still exist
f_close_df.isnull().sum()
```

```python
f_close_df # we have fill that one missing value in the ford dataset
```

```python
import matplotlib.pyplot as plt
import seaborn as sns # Import seaborn for kde plot

sns.histplot(f_close_df['Daily Return'], kde=True )

plt.title('Distribution of Daily Returns')
plt.xlabel('Daily Return')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plotting the close price of Ford
plt.figure(figsize=(10,6))
f_close_df['F_Close'].plot()

#plt.plot(x,y)
plt.title('FORD Close Price')
plt.ylabel('Price')
plt.xlabel('Date')
plt.show()
```

```python
# Plot rolling mean(moving average) and standard deviation
f_close_df['F_Close'].plot(label='Original', figsize=(10,6))
#line plot of original stock prices

f_close_df['F_Close'].rolling(window=30).mean().plot(label='30-Day Moving Average')
#line plot rolling mean values

f_close_df['F_Close'].rolling(window=30).std().plot(label='30-Day Moving Std')
plt.legend()
plt.show()
```

```python
#Handling OUTLIERS
from scipy import stats

#calculate z-scores
z_scores = stats.zscore(f_close_df)

# Define a threshold for outliers (absolute Z-score > 3)
outliers = f_close_df[abs(z_scores) > 3]

# Display the outliers
outliers.dropna()
```

```python
# Define a threshold value for outliers, for example, 5% (0.05)
threshold = 0.05 #significance level

# Use the clip function to limit the 'Daily Return' column to this range
# Any value above 5% is capped to 5%, and any value below -5% is capped to -5%
f_close_df['Daily Return'] = f_close_df['Daily Return'].clip(lower=-threshold, upper=threshold)

# Now, print the statistics of the 'Daily Return' column to see the effect
f_close_df['Daily Return'].describe()
```

### Decomposition of time series,

```python
#decomposition means that to checkj trends and patterns in data
#Stock price ke andar trends(Long term outlook) and patterns hoga(Esa behavior repeated hora ho periodically-daily, quaterly, monthly, semi-anually also called seasonality)
#Repeating patterns at regular interval
```

```python
# Perform seasonal decomposition of the time series
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
f_close_df.index
```

```python
# Decompose the series
decomposition = seasonal_decompose(f_close_df['F_Close'], model="additive", period=12)

# Plot the decomposition results
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
decomposition.plot() #line plots
plt.show()
```

```python
# Decompose the series
decomposition = seasonal_decompose(f_close_df['Daily Return'], model="additive", period=12)

# Plot the decomposition results
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
decomposition.plot() #line plots
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

# Perform the ADF test on the 'F_Close' series
adf_result = adfuller(f_close_df['F_Close'])

# Extracting the results
adf_statistic = adf_result[0]
p_value = adf_result[1]
critical_values = adf_result[4]

adf_statistic, p_value, critical_values
```

#### 


Augmented Dickey-Fuller (ADF) test results are:

ADF Statistic = -3.6848 ,
p-value = 0.0043 ,
Critical Values:
1%: -3.4391 ,
5%: -2.8654 ,
10%: -2.5688 ,
Interpretation:
Compare ADF Statistic with Critical Values:

The ADF Statistic (-3.6848) is lower than all critical values (-3.4391, -2.8654, -2.5688).
This means we can reject the null hypothesis (H₀) at all significance levels (1%, 5%, and 10%).
Check the p-value:

p-value = 0.0043 (< 0.05) → This is significant, meaning we reject H₀.
This confirms that your **'F_Close' time series is stationary**.

#### since our data is already in stationary form , we dont need to check anything like'the Right Order of Differencing'

```python
f_close_df.drop(columns=["Daily Return"], inplace=True)
```

```python
import pmdarima as pm

# Run auto_arima to find the best ARIMA model
auto_model = pm.auto_arima(f_close_df,
                           start_p=0, start_q=0,
                           test='adf',
                           max_p=5, max_q=5,
                           seasonal=False,
                           trace=True,
                           error_action='ignore',
                           suppress_warnings=True,
                           stepwise=True)

# Print the summary of the best model
print(auto_model.summary())

# Forecast the next 30 days using the best ARIMA model
forecast = auto_model.predict(n_periods=30)

# Plot the forecast
plt.figure(figsize=(10, 6))

#line plot of original data or stock prices
plt.plot(f_close_df.index[-100:], f_close_df[-100:], label='Observed')

plt.plot(pd.date_range(f_close_df.index[-1], periods=30, freq='D'), forecast, label='Forecast', color='green')
plt.title('Auto ARIMA 30-Day Forecast')
plt.legend()
plt.show()
```

```python
auto_model.plot_diagnostics(figsize=(10,8))
plt.show()
```

```python
# Fit auto_arima model (with seasonal component)
auto_sarima = pm.auto_arima(f_close_df,
                            seasonal=True,
                            m=12,  # Assuming monthly seasonality
                            start_p=0, start_q=0,
                            test='adf',
                            max_p=3, max_q=3,
                            start_P=0, start_Q=0,
                            max_P=3, max_Q=3,
                            d=1, D=1,
                            trace=True,
                            error_action='ignore',
                            suppress_warnings=True,
                            stepwise=True)

# Print the summary of the best model
print(auto_sarima.summary())

# Forecast the next 30 periods
forecast_sarima = auto_sarima.predict(n_periods=30)

# Plot the observed and forecast values
plt.figure(figsize=(10, 6))
plt.plot(f_close_df.index[-100:], f_close_df[-100:], label='Observed')
plt.plot(pd.date_range(f_close_df.index[-1], periods=30, freq='D'), forecast_sarima, label='Forecast', color='green')
plt.title('Auto SARIMA 30-Day Forecast')
plt.legend()
plt.show()
```
